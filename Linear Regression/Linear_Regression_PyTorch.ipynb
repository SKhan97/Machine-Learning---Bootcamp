```
df = pd.read_csv('airfoil_self_noise.dat', sep = '\t')
df = df.sample(frac = 1) #Reorders data to prevent predictable behaviour
print(df.head())

X = torch.Tensor(np.array(df[df.columns[0:-1]])) #All the features
Y = torch.Tensor(np.array(df[df.columns[-1]]).squeeze()) #Labels 

X =(X-X.mean(0))/X.std(0) #Normalise features such that mean = 0 and s.d. = 0

m = 1100

x_train = Variable(X[0:m])
y_train = Variable(Y[0:m]).view(-1,1) #says make it a column vector

x_test = Variable(X[m:])
y_test = Variable(Y[m:]).view(-1,1)

```

```
class linearmodel(torch.nn.Module):
  def __init__(self):
    super().__init__() #Initialises from the inherited class of torch.nn.Module
    self.linear = torch.nn.Linear(5, 1) #Number of inputs and number of outputs, does a linear combination
  
  def forward(self, x):
    x = self.linear(x);
    return x
```

```
epochs = 1000
lr = 0.5

mymodel = linearmodel()
criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(mymodel.parameters(), lr = lr) #Parameters method inherited from torch.nn.Module
```

```
def train(epochs):
  for epoch in range(epochs):
    h = mymodel.forward(x_train)
    
    cost = criterion(h, y_train)
    #print('Epoch: ', epoch, 'Cost: ', cost.data[0])
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

train(epochs)
```

```
def test():
    h = mymodel.forward(x_test)
    cost = criterion(h, y_test)
    
    return cost.data[0]

test_cost = test()
print('Cost on test set: ', test_cost)
```
